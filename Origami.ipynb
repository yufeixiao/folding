## Real World Problem

How might we program a computer to fold a sheet into real origami that are subject to how paper works?

The mother problem of this question is the field of computer-aided design for metamaterials. If we can reverse engineer the structure of a modular material given our desired properties, then we can make many unintuitive things such as foldable telescopes and stretchy hammers. See examples: [Ruffle stability](https://interactive-structures.org/assets//publications/2021-05-developable-metamaterials/paper.pdf), [Miura cells](https://www.nature.com/articles/s41467-017-00670-w). 

## Computational Formulation

The foldability problem is a kind of global optimization, where the search space contains many local minima (wasteful folds). An optimization problem searches for a set of decision variables that minimizes or maximizes an objective function (target 3D shape) within specified constraints (legal folds). These problems might be encoded as continuous mechanical equations or discrete graphs. The objective function easily grows big and nonlinear because each point adds a dimension to the search space. 

We are using optimization algorithms. The scope of this project is to see which algorithm fits this objective function or general problem better. after all, there is no free lunch (no optimization algorithm outperforms another on average benchmarks), without the context of the problem there is not one superior algorithm that can rule all problems. it's the uniqueness of place.

In this project I implement two optimization algorithms in Python to solve optimization problems with such hilly landscapes.


To summarize the problem: global optimization over a nonconvex continuous objective function.

- Inputs: domain of possible solutions, objective function and constraint

- Optional inputs for testing: global optima, error tolerance

- Outputs: best solution to the objective function, its value

A note on convergence to nonlinear objectives: Imagine the search space as an energy landscape, then the global minima is the equilibrium steady state. Solvers might not get to the global minima. When the search space is big (every vertex has one dimension), we can only approximate the objective function. That is why to meaningfully analyze our solutions, we had to set more toleratnt thresholds on iterations and error to terminate the program.

## Proposed Solutions

Solutions: derivative-based (gradient descent) and derivative-free (particle swarm optimization). 

- Gradient Descent special inputs: gradient of the cost function, learning rate

- Particle Swarm Optimization special inputs: inertial weight, velocity coefficients

- swarm size N

- exploration vs exploitation through parameter tuning

## Gradient Descent

Pseudocode
 from [Madni 2019](https://www.researchgate.net/figure/Pseudo-code-of-hybrid-gradient-descent-cuckoo-search-HGDCS-algorithm_fig6_328464831)

![image](https://user-images.githubusercontent.com/51347874/115949073-42defd00-a50d-11eb-8b69-ee0ef3534c25.png)


Gradient descent is a greedy algorithm that takes steps to the direction of largest height difference. It requires the derivative of the objective function. 

```python
class GD:

  """
  performs gradient descent optimization on differentiable functions.

  PARAMETERS
  ---
  required:
  - objective: cost function
  - gradient: first derivative of cost function
  - dim: dimension of solutions

  optional:
  - learnrate: step size
  - maxiter: maximum iterations
  - precision: tolerable step size
  - lb: lower bound for solutions
  - ub: upper bound for solutions

  RETURNS
  ---
  - iters, values: array of objective value for each step

  """

  # TODO: optimize `learnrate` using global optima
  def __init__(self, objective, gradient, \
               dim, learnrate=0.01, maxiter=100, precision=None, \
               lb=None, ub=None, iters=[], values=[]):
    
    self.objective = objective
    self.gradient = gradient
    self.learnrate = learnrate
    self.maxiter = maxiter # max iteration
    self.precision = precision # tolerable stepsize

    # initialize the position array
    self.positions = np.zeros(shape=(maxiter+1, dim))

    # randomize the starting position
    if lb and ub: 
      start = np.random.uniform(low=lb, high=ub, size=dim)
    else:
      start = np.zeros(shape=dim)

    self.positions[0] = start
    self.step = 0

    self.iters = iters
    self.values = values

  # take one step using size learnrate and direction gradient
  def move(self, iter):

    self.step = self.gradient(self.positions[iter]) * self.learnrate
    self.positions[iter+1] = self.positions[iter] - self.step
    
  # main loop of gradient descent
  def optimize(self):

    # store steps for plotting
    iters, values = [], []

    # terminate when max iterations is reached
    for iter in range(self.maxiter): 

      # descend and store one step
      self.move(iter)
      self.iters.append(iter)
      self.values.append(self.objective(self.positions[iter]))

      # terminate when good stepsize is reached
      if self.precision:
        if np.abs(self.step) <= self.precision:
          break

    return self.iters, self.values
```

Gradient descent does not guarantee global optima in non convex! because it might get stuck in a small pond A lower learning rate prevents the vector from making large jumps. Besides the learning rate, the starting point can affect the solution significantly especially with nonconvex functions.

## Particle Swarm Optimization

![image](https://user-images.githubusercontent.com/51347874/115944630-ddc9de00-a4f1-11eb-8f14-f3ed0acc895d.png)

Pseudocode from (Chen, 2010), where: $\mathbb{F}$ is fitness function, $Pb$ is personal best position, $Nb$ is global best position, $V$ is velocity, $X$ is position, $t$ is timestep, $\otimes$ is component-by-component multiplication.

```python
class Particle:

  def __init__(self, dim, objective, lb=-5, ub=5):

    self.objective = objective
    self.position = np.random.uniform(low=lb, high=ub, size=dim)
    self.value = self.objective(self.position)
    self.velocity = np.random.uniform(low=lb, high=ub, size=dim)
    self.pbest_pos = self.position.copy()
    self.pbest_val = self.objective(self.pbest_pos)
```

```python
class Swarm:

  def __init__(self, dim, objective, iters=[], values=[], \
               lb=-5, ub=5, w=0.8, c1=1, c2=1, swarmsize=100, maxiter=100):
    self.objective = objective
    self.particles = [Particle(dim=dim,objective=objective,lb=lb,ub=ub) for _ in range(swarmsize)]
    
    self.swarmsize = swarmsize
    self.w = w
    self.c1 = c1
    self.c2 = c2
    self.maxiter = maxiter

    self.gbest_pos = np.random.choice(self.particles).position
    self.gbest_val = self.objective(self.gbest_pos)
    self.update_bests()

    self.iters = iters
    self.values = values

  # updates velocity and position of all particles in one iteration
  def move(self):
    for particle in self.particles:
      # add inertial weight
      new_vel = self.w * particle.velocity
      # add cognitive weight
      r1 = np.random.random()
      new_vel += self.c1 * r1 * (particle.pbest_pos - particle.position)
      # add social weight
      r2 = np.random.random()
      new_vel += self.c2 * r2 * (self.gbest_pos - particle.position)
      # update position and velocity
      particle.velocity = new_vel
      particle.position = particle.position + new_vel
      # TODO: check stagnancy by setting a boolean: 
      # self.is_running = np.all(np.sum(particle.velocity - new_vel) != 0)

  # updates best personal and global positions 
  def update_bests(self):

    # iterates through the array of particles
    for particle in self.particles:
      particle.value = self.objective(particle.position)
      # update pbest (cognitive)
      if particle.value < particle.pbest_val:
        particle.pbest_val = particle.value
        particle.pbest_pos = particle.position
        # update gbest (social)
        if particle.value < self.gbest_val:
          self.gbest_val = particle.value
          self.gbest_pos = particle.position

  # wrapper for running PSO
  def optimize(self):

    # terminates when max iterations reached
    for iter in range(self.maxiter):
      # move particles
      self.move()
      # re-orient to two attractors
      self.update_bests()
      # store values
      self.iters.append(iter)
      self.values.append(self.gbest_val)

    return self.iters, self.values
```

# Analysis

## Theoretical analysis

To approach the complexity of optimization algorithms we might consider such properties as the number of operations per step, progress rate, convergence probability, ease of getting stuck. I will start by looking at the structure of the pseudocode.

Gradient descent complexity might be approximated by cost of update = (Cost of gradient + cost of evaluating OF) * number of updates.

PSO might be approximated similarly by Number of particles * cost of update * number of updates. 

We set controlled cutoffs such as max number of steps and the dimensions of the problem. Both scales with the dimensions of the problem, indicated by the number of components in the vector. Our benchmarks are 2D, but scaling up I imagine the gradient computation becomes more costly than the component-wise multiplication of PSO.


## Empirical tests

These are continuous solvers. To simulate, we would discretion space and time (learning rate). We would convert noisy behavioral functions in the PSO into transition rules that approximate the continuous model. 

```python
def test(objective, gradient, dim, nsims, maxiter=100):

  """
  applies GD and PSO solvers on a target differentiable objective.
  outputs a plot of accuracy over iterations.
  """

  #store data
  x1, y1 = [], []
  x2, y2 = [], []

  #simulate
  for sim in range(nsims):

    #GD
    gd = GD(objective=objective, gradient=gradient, dim=dim, \
            learnrate = 0.01, maxiter=maxiter, iters=x1, values=y1)
    gd.optimize()
    #PSO
    pso = Swarm(objective=objective, dim=dim, maxiter=maxiter, \
                iters=x2, values=y2)
    pso.optimize()

  #plot solutions
  ax=sns.lineplot(x1,y1,label="GD")
  ax=sns.lineplot(x2,y2,label="PSO")
  ax.set(yscale="log")
  plt.ylabel('log error')
  plt.xlabel('iterations')
  plt.legend()

def plot_objective(objective):

  """
  plots an objective function with dim=2 inputs.
  """

  # Initialize figure 
  fig = plt.figure(figsize=(12, 7))
  ax = fig.gca(projection='3d')

  # Evaluate function
  X = np.arange(-4, 4, 0.15)
  Y = np.arange(-4, 4, 0.15)
  X, Y = np.meshgrid(X, Y)
  Z = objective([X,Y])

  # Plot the surface
  surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.PuOr,
                        linewidth=0, antialiased=False)
  fig.colorbar(surf, shrink=0.5, aspect=10)
  plt.title(f'{objective}')
  plt.show()

  # Plot the contour
  fig, ax = plt.subplots()
  ax.contour(X,Y,Z,15,colors='k',linewidths=0.5)
  ax.contourf(X,Y,Z,50,cmap=plt.cm.PuOr)
  plt.title(f'{objective}')
```

I tested two common global optimization benchmarks: sphere function and rosenbrock function. Both are continuous and differentiable even though PSO does not require differentiability. Rosenbrock function is nonconvex, meaning there exists local minima. It is expected to trap the gradient solver, because the solver stops when stepsize shrinks below the precision limit, which likely occurs at the shallow basins.

```python
# test 1: Sphere
#[lb, ub] = np.array([[-5.12, 5.12] for _ in range(dim)])

def sphere(pos):
  x, y = pos[0], pos[1]
  return (x**2 + y**2)

def grad_sphere(pos):
  x, y = pos[0], pos[1]
  return np.array([2*x,2*y])

# visualize objective landscapes
plot_objective(rosenbrock)
plot_objective(sphere)

# test 2: rosenbrock. bad landscape for Gradient Descent
# formula: 'f(\mathbf{x})=\sum_{i=1}^{d-1}[b (x_{i+1} - x_i^2)^ 2 + (a - x_i)^2]'
# domain: x_i \in [-5, 10], \forall i \in \llbracket 1, d\rrbracket
# global min:  'f(1, ..., 1)=0'

def rosenbrock(pos):
  x, y = pos[0], pos[1]
  return (x ** 2 - 10 * np.cos(2 * np.pi * x)) + \
           (y ** 2 - 10 * np.cos(2 * np.pi * y)) + 20
def grad_rosenbrock(pos):
  x, y = pos[0], pos[1]
  return np.array([-2 * x + 10 * 2 * np.pi * x * np.sin(2 * np.pi * x),
         2 * y + 10 * 2 * np.pi * y * np.sin(2 * np.pi * y)])
  
```

In testing an important consideration is that the solvers might not reach global optimum. therefore, we need to 'relax' the goal a little to have convergence / termination, such as a threshold over the error or a limit on the number of steps. I plot the accuracy vs number of iterations to compare the two algorithms with the same learning rate, but a more controlled method might be to compare relative error over error tolerance while other parameters are comparable.

```python
test(objective=sphere, gradient=grad_sphere, dim=2, nsims=10, maxiter=100)
```

```python
# run tests
test(objective=rosenbrock, gradient=grad_rosenbrock, dim=2, nsims=10, maxiter=100)
```

There are still tradeoffs of particle and swarm efficiency. If a process is on expectation decreasing, then drift theory allows bounds on expected time until hit 0. The stochastic interaction between the attractor gbest and the individual particles that sample from phase space lets them help each other, but sometimes PSO does not converge: for example, in some coefficients of the Rosenbrock function the swarm stops randomly at a local optima or a non-optimal point, because it terminates at max iterations in a huge dimensional space or it happens to all come to velocity zero and cluster together. Then the hitting time to our desired neighborhood has infinite expectation time which is bad. To improve we might add perturbations, or more reasonable domains, or hybridize the algorithm with more explorative walkers that escape local.

### PSO Parameter Tuning

A challenge in heuristic optimizer design is tuning the parameters to fit the problem. The parameters of PSO have physical grounding: they correspond to the particle bird's inertial weight, social interaction, memory. Researchers find tight bounds on the parameters using the large N approximation to guarantee convergence in finite time. Further, they add parameters to model real swarms and remove PSO failure modes such as getting stuck fast in the locals.

I tried some tweaks to see what happens to the swarm state.

First I took the standard fixed parameters experimentally derived by Clerc and Kennedy. 

Fixed parameters: 
$$w=0.729\\c1=1.49\\c2=1.49\\lr=0.01$$

In conclusion, vanilla PSO outperforms vanilla gradient descent in two standard global optimization benchmarks in terms of cost (iterations as a proxy for time) and accuracy (error from the global minima). 

There are many data structures and algorithmic improvements that are possible. For gradient descent we can definitely give it randomness and varying step sizes to overcome barriers. Even in the Spherical benchmark problem, we did not get so close to optima and we manually tuned parameters. The biggest failure mode of PSO is stagnation. We might alter the neighborhood shape to have more interactions and learning from the less than best neighbors to get more fluctuations.
